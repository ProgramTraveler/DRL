# Save and load the Model

* we will look at how to persist model state with saving, loading and run model predictions.

```python
import torch
import torchvision.models as models
```

## Saving and loading Model weights

* **PyTorch** models store the learning parameters in an internal state dictionary, called `state_dict`. These can be persisted via the `torch.save` method.

```python
model = models.vgg16(weights='IMAGENET1K_V1')
torch.save(model.state_dict(), 'model_weight.pth')
```

* To load model weights, you need to create an instance of the same model first, and then load the parameters using `load_state_dict()` method.

```python
model = models.vgg16() # we do not specify ``weight``, i.e. create untrained model
model.load_state_dict(torch.load('model_weight.pth'))
model.eval()
```

* **NOTE:** be sure call `model.eval()` method before inferencing to set the dropout and batch normalization layers to evaluation mode. Failing to do this will yield inconsistent inference results.

```python
torch.save(model, 'model.pth')
```

* we can then load the model like this:

```python
model = torch.load('model.pth')
```

* **NOTO:** This approach uses Python pickle module when serializing the model, thus it relies on the actual class definition to be available when loading the model.
